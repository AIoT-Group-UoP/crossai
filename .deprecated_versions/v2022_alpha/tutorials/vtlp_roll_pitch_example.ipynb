# Load Instance

Use a package that can load the .wav files as an NumPy array (such as *librosa*)

```python
# audio instance path (tutorials/data_test folder)
data_path = os.path.join(os.getcwd(), "data_test", "audio.wav")
```

```python
# info about the instance
audio, s = librosa.load(data_path)
print("Type of audio", type(audio))
print("Shape of audio:", audio.shape)
print("Sampling rate:", s)
```

```python
data = [{'X': audio, 'Y': "hiphop"}]
```

```python
from crossai.processing import Audio
crossai_audio = Audio(data)
crossai_audio
```

```python
crossai_audio.data[0]['X'][0].shape
```

```python
original_data = crossai_audio.data[0]['X'][0]
original_data
```

```python
crossai_audio.vtlp_aug(sr = s)
crossai_audio.data
```

```python
crossai_audio.PitchAug(sr = s, factor = (1, 3))
crossai_audio.data
```

```python
crossai_audio.RollAug()
crossai_audio.data
```

```python
crossai_audio.data[0]['X'][0].shape
```

```python

plt.figure(figsize=(20, 10))
plt.title("Pitch Augmentation")
plt.plot(original_data, label="original")
plt.plot(crossai_audio.data[0]['X'][0][0], label="augmented")
plt.legend()
plt.show()
```

```python

plt.figure(figsize=(20, 10))
plt.title("Vtlp Augmentation")
plt.plot(original_data, label="original")
plt.plot(crossai_audio.data[0]['X'][0][0], label="augmented")
plt.legend()
plt.show()
```
